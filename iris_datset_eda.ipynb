```python
import numpy as np
iris_setosa = iris.loc[iris['variety'] == 'Setosa']
iris_versicolor = iris.loc[iris['variety'] == 'Versicolor']
iris_virginica = iris.loc[iris['variety'] == 'Virginica']
#print(iris_setosa["petal.length"])
plt.plot(iris_setosa['petal.length'], np.zeros_like(iris_setosa['petal.length']),'o')
plt.plot(iris_versicolor["petal.length"], np.zeros_like(iris_versicolor['petal.length']), 'o')
plt.plot(iris_virginica["petal.length"], np.zeros_like(iris_virginica['petal.length']), 'o')
plt.show()
```

```python
import numpy as np
counts, bin_edges = np.histogram(iris_setosa['petal.length'],bins=10,density=True)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
import numpy as np
counts, bin_edges = np.histogram(iris_setosa['petal.length'],bins=10,density=True)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

counts, bin_edges = np.histogram(iris_versicolor['petal.length'],bins=10,density=True)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

counts, bin_edges = np.histogram(iris_virginica['petal.length'],bins=10,density=True)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
print('mean\n')
l = np.mean(iris_setosa['petal.length'])
u = np.mean(iris_versicolor['petal.length'])
x = np.mean(np.append(iris_versicolor['petal.length'],50))
z = np.mean(iris_virginica['petal.length'])
print(l)
print(x)
print(u)
print(z)

print('\nstd_dev\n')
l = np.std(iris_setosa['petal.length'])
u = np.std(iris_versicolor['petal.length'])
z = np.std(iris_virginica['petal.length'])
print(l)
print(u)
print(z)
```

```python
print('\nmedian\n')
l = np.median(iris_setosa['petal.length'])
u = np.median(iris_versicolor['petal.length'])
z = np.median(iris_virginica['petal.length'])
print(l)
print(u)
print(z)

print('\npercentile\n')
l = np.percentile(iris_setosa['petal.length'],np.arange(0,100,25))
u = np.percentile(iris_versicolor['petal.length'],np.arange(0,100,25))
z = np.percentile(iris_virginica['petal.length'],np.arange(0,100,25))
print(l)
print(u)
print(z)

print('\n90th percentile\n')
l = np.percentile(iris_setosa['petal.length'],90)
u = np.percentile(iris_versicolor['petal.length'],90)
z = np.percentile(iris_virginica['petal.length'],90)
print(l)
print(u)
print(z)

print('\nmad\n')
from statsmodels import robust
print(robust.mad(iris_setosa['petal.length']))
print(robust.mad(iris_versicolor['petal.length']))
print(robust.mad(iris_virginica['petal.length']))
```
